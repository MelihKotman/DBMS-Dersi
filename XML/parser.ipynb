```python
# lxml kütüphanesinden etree (ElementTree) modülünü içe aktarıyoruz.
# Bu modül, XML dosyasını tıpkı bir soyağacı gibi düğümlere (nodes) ayırmamızı sağlar.
from lxml import etree
```

```python
# 1. PARSER (AYRIŞTIRICI) VE DTD DOĞRULAMASI
# Hocanın 16. sayfadaki notu: "DTD'ye uygun olması için dtd_validation=True açtık" .
# Sadece XML'i okumakla kalmıyoruz, aynı zamanda daha önce yazdığımız o katı DTD kurallarına 
# (Örn: yazar zorunlu mu? id girilmiş mi?) uyup uymadığını kontrol eden bir denetmen (parser) yaratıyoruz.
parser = etree.XMLParser(dtd_validation=True)
```

```python
# "katalog.xml" dosyasını okuyoruz ve yukarıda yarattığımız denetmenden geçiriyoruz.
# Eğer XML dosyası DTD kurallarını ihlal ediyorsa, program tam bu satırda hata (Exception) fırlatır ve durur.
# Geçerse, 'tree' adında sanal bir ağaç objesi oluşur.
tree = etree.parse("katalog.xml", parser)
```

```python
# 2. KÖK ELEMANA (ROOT) ULAŞMAK
# Ağacın en tepesine çıkıyoruz. Bizim senaryomuzda bu <katalog> etiketidir.
root = tree.getroot()
```

```python
# Kök elemanın doğrudan altında ne var? (Yani <kitaplar> ve <plaklar> elemanları)
# Döngü ile çocukları gezip, etiket adlarını (tag) ve varsa niteliklerini (attrib) ekrana yazdırıyoruz.
for child in root:
    print("Kökün Çocuğu:", child.tag, "| Nitelikleri:", child.attrib)
```

```python
# Kök elemanın kendi adını ("katalog") yazdırır.
print("Kök Adı:", root.tag)

# Kök elemanın kaç tane çocuğu olduğunu yazdırır. 
# Eğer sadece <kitaplar> ve <plaklar> varsa çıktı '2' olacaktır.
print("Kökün Çocuk Sayısı:", len(root))
```

```python
# 3. İNDEKSLERLE AĞAÇTA GEZİNME (Aşağı Doğru İniş)
# XML ağacındaki elemanlara diziler (arrays) gibi [0], [1] indeksleriyle erişebiliriz.
# root[1] demek, kök elemanın 2. çocuğu demektir. Bizim DTD'ye göre bu <plaklar> düğümüdür.
plaklar = root[1]

# <plaklar> düğümünün içindeki her bir <plak> elemanını tek tek dönüyoruz.
for plak in plaklar:
    print("Plak Etiketi:", plak.tag, "| Plak Nitelikleri:", plak.attrib)


# root[0] demek, kök elemanın 1. çocuğu demektir. Bu da <kitaplar> düğümüdür.
kitaplar = root[0]

# <kitaplar> düğümünün içindeki her bir <kitap> elemanını tek tek dönüyoruz.
for kitap in kitaplar:
    print("Kitap Etiketi:", kitap.tag, "| Kitap Nitelikleri:", kitap.attrib)
```

```python
# 4. SPESİFİK BİR NİTELİĞİ (ATTRIBUTE) ÇEKME
# <kitaplar> düğümünün içindeki 1. kitabı (indeks 0) seçip bir değişkene atıyoruz.
ilk_kitap = kitaplar[0]

# Seçtiğimiz bu ilk kitabın etiket adını yazdırıyoruz (Çıktı: "kitap").
print("Seçilen Eleman:", ilk_kitap.tag)

# DTD'de "#REQUIRED" (Zorunlu) kıldığımız 'id' niteliğini çekiyoruz.
# .attrib yapısı Python'da bir sözlük (dictionary) döndürür. ["id"] diyerek doğrudan o değere (Örn: "bk1") ulaşıyoruz.
# print("İlk Kitabın ID'si:", ilk_kitap.attrib["id"])
```
